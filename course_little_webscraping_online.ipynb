```python
paragraphs = soup.find_all('p')
#print(paragraphs)
meaningful_text = []
for one_p in paragraphs:
    whats_under_the_tag = one_p.contents
    what_to_append = whats_under_the_tag[0].replace('\r\n', '')
    what_to_append = what_to_append.replace('\xa0', ' ')
    #print(what_to_append)
    #break
    meaningful_text.append(what_to_append)
print(meaningful_text)
```

```python
# all chapters:
book_name = 'alice'
CORRECT_RESPONSE = 200
chapters = []
chapters_contents = []
for page_num in range(1, 1000):
    url = f'http://www.literatureproject.com/{book_name}/{book_name}_{page_num}.htm'
    try:
        response = requests.get(url, timeout = 100)
    except:
        break
    if response.status_code != CORRECT_RESPONSE:
        break
    soup = BeautifulSoup(response.content, 'html.parser')
    chapter_title = soup.title.contents
    chapters.append(chapter_title)
    paragraphs = soup.find_all('p')
    # print(paragraphs)
    meaningful_text = []
    for one_p in paragraphs:
        whats_under_the_tag = one_p.contents
        what_to_append = whats_under_the_tag[0].replace('\r\n', '')
        what_to_append = what_to_append.replace('\xa0', ' ')
        #print(what_to_append)
        #break
        meaningful_text.append(what_to_append)
    #print(meaningful_text)
    chapters_contents.append(meaningful_text)
```

```python
file = open("whole_text_eng.txt", "w")
```

```python
for chap_num in range(len(chapters)):
    file.write(chapters[chap_num][0])
    one_chap = chapters_contents[0]
    for one_paragraph in one_chap:
        file.write(one_paragraph)
```
